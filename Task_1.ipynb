```python
# Function to cap outliers in a DataFrame
def cap_outliers_in_dataframe(train_data_imputed_mean, lower_percentile, upper_percentile):
    """
    Caps the outliers in all numerical columns of a DataFrame at specified percentiles.
    
    Parameters:
    df (DataFrame): The DataFrame to process.
    lower_percentile (float): The lower percentile to use as the cap for lower outliers.
    upper_percentile (float): The upper percentile to use as the cap for upper outliers.

    Returns:
    DataFrame: The DataFrame with outliers capped.
    """

    # We create a new DataFrame to avoid modifying the original one
    train_data_imputed_mean_capped = train_data_imputed_mean.iloc[:, 2:].copy()

    # Iterate over all columns in the DataFrame
    for column in train_data_imputed_mean_capped.columns:
        # Check if the column is numerical
        if pd.api.types.is_numeric_dtype(train_data_imputed_mean_capped[column]):
            # Calculate the percentiles for this column
            lower_bound = train_data_imputed_mean_capped[column].quantile(lower_percentile)
            upper_bound = train_data_imputed_mean_capped[column].quantile(upper_percentile)

            # Cap the values using the clip method
            train_data_imputed_mean_capped[column] = train_data_imputed_mean_capped[column].clip(lower=lower_bound, upper=upper_bound)

    return train_data_imputed_mean_capped

# Assume 'df' is your DataFrame, and you want to use the 5th and 95th percentiles
lower_percentile = 0.05
upper_percentile = 0.95

# Cap outliers in the entire DataFrame
train_data_imputed_mean_capped = cap_outliers_in_dataframe(train_data_imputed_mean, lower_percentile, upper_percentile)

train_data_imputed_mean_capped
```
